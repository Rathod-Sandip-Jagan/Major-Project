```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

file_path = r'/content/Full_data.csv'

try:
    with open(file_path, 'r') as csvfile:
        lines = csv.reader(csvfile, delimiter=',')
        next(lines) #skip header
        for row in lines:
            x.append(row[0])
            y.append(int(row[32]))

    plt.rcParams["figure.figsize"] = (50, 50)
    plt.plot(x, y, color='g', linestyle='dashed', marker='o', label="Potato Price")

    plt.xticks(rotation=100)
    plt.xlabel('Date')
    plt.ylabel('Potato Price (Per Kg)')
    plt.title('Potato Price', fontsize=20)
    plt.grid()
    plt.legend()
    plt.show()
    plt.plot(x, y, color='g', linestyle='dashed', marker='o', label="Onion Price")

    plt.rcParams["figure.figsize"] = (150, 50)
    plt.xticks(rotation=100)
    plt.xlabel('Date')
    plt.ylabel('Onion Price (Per Kg)')
    plt.title('Onion Price', fontsize=20)
    plt.grid()
    plt.legend()
    plt.show()
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
```
